```python
import pandas as pd
from ipyvizzu import Config, Data, Style
from ipyvizzustory import Story, Slide, Step

df = pd.read_excel("data/dataset.xlsx")
df["aux_index"] = df.index.copy().astype("str")
df.head()
```

```python
data = Data()
data.add_df(df)

story = Story(data)
story.set_size("800px", "800px")
my_style = Style({
                "plot": {
                    "marker": {"colorPalette": "#ff0000 #0000ff"}
                }
            })

slide = Slide(Step(Config({
            "x": "group",
            "y": "count()",
            "color": "group",
        }),
        my_style))

story.add_slide(slide)  

slide = Slide(Step(Config({
            "x": "x",
            "y": "y",
            "color": "aux_index",
            "title": "Python Chile según Vizzu",
            "geometry": "circle",
            }),
        ))

story.add_slide(slide)

story.play()
```
